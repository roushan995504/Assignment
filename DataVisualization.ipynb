```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot data to create a matrix of sales with Days as rows and Months as columns
pivot_df = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_df, cmap='coolwarm', annot=True, fmt=".0f")
plt.title("Sales Variation across Days and Months")
plt.xlabel("Month")
plt.ylabel("Day")
plt.show()
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Ensure output to Jupyter Notebook
output_notebook()

# Generate sine wave data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the Bokeh plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label='sin(x)', line_width=2)

# Show the plot
show(p)

```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.


from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Ensure output to Jupyter Notebook
output_notebook()

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.rand(100) * 50 + 10  # Adjust for better visibility
colors = np.random.randint(0, 255, 100)  # Random color indices

# Create the Bokeh scatter plot
p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')
p.scatter(x, y, size=sizes, color="blue", alpha=0.6)

# Show the plot
show(p)



```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

output_notebook()
# Data for bar chart
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts')
p.vbar(x='fruits', top='counts', width=0.5, source=source, color="navy")

# Show the plot
show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
from bokeh.plotting import figure, show
import numpy as np
output_notebook()
# Generate random data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a histogram plot
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="white")

# Show the plot
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
import numpy as np
output_notebook()
# Generate heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a figure
p = figure(title="Heatmap", x_axis_label='x', y_axis_label='y', toolbar_location=None)

# Create color mapper
mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Add a heatmap image
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=mapper)

# Add color bar
color_bar = ColorBar(color_mapper=mapper, label_standoff=12, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
